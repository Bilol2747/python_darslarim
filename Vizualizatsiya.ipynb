```python
x=np.arange(0, 4*np.pi, 0.1)
cos=np.cos(x)
sin=np.sin(x)
plt.figure(figsize=(10, 4))
plt.plot(x, sin, linewidth=1.5, linestyle='--', color='r', marker='>')
plt.title('Sin(x) funksiya')
plt.xlabel('Pi')
plt.ylabel('Sin(x)')
plt.grid(linewidth=1, linestyle='--', color='b')

#plt.figure(figsize=(10, 4))
plt.plot(x, y, linewidth=1.5, linestyle='--', color='b', marker='>')
plt.title('Cos(x) funksiya')
plt.xlabel('Pi')
plt.ylabel('Cos(x)')
plt.grid(linewidth=1, linestyle='--', color='r')
plt.show()
```
