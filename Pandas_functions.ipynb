```python
df=pd.read_csv('/content/Practicum_Datasets/world_cities_air_water_quality.csv')
df
```

```python
column=['City', 'Region', 'Country', 'AirQuality', 'WaterPollution']
df.columns=column
```

```python
df[['Region', 'Country']]=df[['Region', 'Country']].map(lambda x: x.replace('"', ""))
df
```

```python
df.to_csv("Updated_df", index=False)
```

```python
df['Country'][1]
```

```python
df['Country']=df['Country'].map(lambda x: x.strip())
```

```python
df.head()
```

```python
df1=pd.pivot_table(df, index=['Country', 'City'], values=['AirQuality', 'WaterPollution'])
```

```python
print(df1.head(100))
```

```python
df1.loc[['Uzbekistan', 'Tajikistan']]
```

```python
df2=df1.copy()
```

```python
df2
```

```python
df2['AirPollution'] = 100-df2['AirQuality']
df2['WaterQuality'] = 100-df2['WaterPollution']
df2
```

```python
df3=pd.DataFrame(df2.values, index=df2.index, columns=[['Air', 'Air', 'Water', 'Water'], ['AirQuality', 'AirPollution', 'WaterPollution', 'WaterQuality']])
```
