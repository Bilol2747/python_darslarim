```python
df=pd.read_csv("https://github.com/anvarnarz/praktikum_datasets/raw/main/museum_visitors.csv", index_col=0, parse_dates, usecols=[0, 1])
df.head()
```

```python
type(df.index)
```

```python
df['month']= [i.month for i in df.index]
df.head()
```

```python
df['year']=[i.year for i in df.index]
df.head()
```

```python
df2=df.groupby(['month', 'year']).max()
df2
```

```python
df2.unstack(level=1)
```

```python
import matplotlib.pyplot as plt
%matplotlib inline
import seaborn as sns
```

```python
plt.figure(figsize=(12,6))
sns.heatmap(data=df2.unstack(level=0), cmap='Accent', linewidths=2, square=True)
plt.show()
```

```python
import numpy as np
import pandas as pd
```

```python
import matplotlib.pyplot as plt
import seaborn as sns
%matplotlib inline
```

```python
df=pd.read_csv("https://raw.githubusercontent.com/anvarnarz/praktikum_datasets/main/merc.csv")
df.head()
```

```python
df.shape
```

```python
df=df[df['year']>2000]
plt.figure(figsize=(10,6))
sns.scatterplot(data=df, x='year', y='price')
plt.show()
```

```python
df=df[df['year']>2000]
plt.figure(figsize=(10,6))
##sns.scatterplot(data=df, x='year', y='price')
sns.regplot(data=df, x='year', y='price')
plt.show()
```

```python
df=df[df['year']>2000]
plt.figure(figsize=(16,8))
sns.scatterplot(data=df, x='mileage', y='price', style='fuelType', hue= 'transmission', s=80)
#sns.regplot(data=df, x='mileage', y='price')
plt.show()
```

```python
df=pd.read_csv("https://raw.githubusercontent.com/anvarnarz/praktikum_datasets/main/merc.csv")
df.head()
```

```python
df=df[df.year>2010]
df=df[df.transmission!='other']
df=df[df.fuelType!='other']
df.head()
```

```python
plt.figure(figsize=(10,6))
sns.histplot(data=df, x='price')
plt.show()
```

```python

sns.displot(data=df, x='price', height=5, hue='transmission',aspect=2)
plt.show()
```

```python
sns.displot(data=df, x='price', height=8, col='transmission',aspect=2)
plt.show()
```

```python
sns.displot(data=df, x='price', kind='kde', height=5, aspect=2)
plt.show()
```

```python
sns.displot(data=df, x='price', kde=True, hue='fuelType', height=5, aspect=2)
plt.show()
```
