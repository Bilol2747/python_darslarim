#2-topshiriq: Series o'rnatish
Quyidagi raqamlardan iborat Series ob'ektini yarating: 5, 10, 15, 20, 25. Har bir elementga A, B, C, D, E harflari ko'rinishidagi indeks belgilang. Yaratilgan qatorlarni ekranda ko'rsatish.

```python
df1 = pd.DataFrame({'ProductID': [1, 2, 3], 'Name': ['Яблоко', 'Банан', 'Виноград']})
df2 = pd.DataFrame({'ProductID': [1, 2, 3], 'Price': [100, 200, 300]})
```
