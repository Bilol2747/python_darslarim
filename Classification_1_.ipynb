```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
%matplotlib inline
```

```python
df = pd.read_csv("https://raw.githubusercontent.com/anvarnarz/praktikum_datasets/main/breast-cancer.csv")
df.head(10)
```

```python
df.shape
```

```python
df.info()
```

```python
df['diagnosis'].value_counts()
```

```python
from sklearn.preprocessing import LabelEncoder
labelEncoder= LabelEncoder()
df['diagnosis'] = labelEncoder.fit_transform(df['diagnosis'].values)
df['diagnosis'].value_counts()
```

```python
df.drop(columns='id', inplace=True)
```

```python
df_cor=df.corr().abs()
df_cor.style.background_gradient(cmap='coolwarm')
```

```python
df.corrwith(df['diagnosis']).abs().sort_values(ascending=False)
```

```python
x=df.drop('diagnosis', axis=1).values
```

```python
y=df['diagnosis']
```

```python
from sklearn.preprocessing import StandardScaler
scaler=StandardScaler()
x= scaler.fit_transform(x)
```

```python
from sklearn.model_selection import train_test_split
x_t, x_s, y_t, y_s= train_test_split(x, y, test_size=0.2, random_state=45)
```

```python
from sklearn.neighbors import KNeighborsClassifier
knn=KNeighborsClassifier(n_neighbors=1)
knn.fit(x_t, y_t)
y_predict= knn.predict(x_s)
```

```python
from sklearn.metrics import jaccard_score
jaccard_score(y_s, y_predict)
```

```python
from sklearn.metrics import confusion_matrix
sns.heatmap(confusion_matrix(y_s, y_predict), annot=True)
plt.show()
```

```python
confusion_matrix(y_s, y_predict)
```

```python
from sklearn.metrics import precision_score, accuracy_score, f1_score, recall_score
precision_score(y_s, y_predict)
```

```python
from sklearn.metrics import classification_report
print(classification_report(y_s, y_predict))
```

```python
from sklearn.model_selection import cross_val_predict
predict = cross_val_predict(estimator=knn, X=x, y=y, cv=5)
print(classification_report(y, predict))
```

```python
f1=[]
for k in range(1, 25):
  knn= KNeighborsClassifier(n_neighbors=k)
  knn.fit(x_t,  y_t)
  y_predict=knn.predict(x_s)
  f1.append(f1_score(y_s,  y_predict))
plt.figure(figsize=(10, 5))
plt.plot( range(1, 25), f1)
plt.xticks(range(1, 25))
plt.grid()
plt.show()
```
